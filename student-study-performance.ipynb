```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/kaggle/input/student-study-performance/study_performance.csv')

# Gender
gender_counts = df['gender'].value_counts()
print('Gender Counts:\n', gender_counts)
```

```python
# Race/Ethnicity
race_ethnicity_counts = df['race_ethnicity'].value_counts()
print('\nRace/Ethnicity Counts:\n', race_ethnicity_counts)
```

```python
# Parental Level of Education
parental_education_counts = df['parental_level_of_education'].value_counts()
print('\nParental Level of Education Counts:\n', parental_education_counts)
```

```python
# Lunch
lunch_counts = df['lunch'].value_counts()
print('\nLunch Counts:\n', lunch_counts)
```

```python
# Test Preparation Course
test_prep_counts = df['test_preparation_course'].value_counts()
print('\nTest Preparation Course Counts:\n', test_prep_counts)
```

```python
# Math Score
math_score_counts = df['math_score'].value_counts().sort_index()
print('\nMath Score Counts:\n', math_score_counts.head())
```

```python
# Reading Score
reading_score_counts = df['reading_score'].value_counts().sort_index()
print('\nReading Score Counts:\n', reading_score_counts.head())
```

```python
# Writing Score
writing_score_counts = df['writing_score'].value_counts().sort_index()
print('\nWriting Score Counts:\n', writing_score_counts.head())
```

```python
# The distribution of each category within these first 10 entries.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Consider the first 10 rows of the dataset for visualization
df_head_10 = df.head(10)

# Plotting each categorical variable
categorical_columns = ['gender', 'race_ethnicity', 'parental_level_of_education', 'lunch', 'test_preparation_course']
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=column, data=df_head_10)
    plt.title(f'Count of {column.capitalize()} in First 10 Rows')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# The subjects scores exams for each of the first 10 entries
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Consider the first 10 rows of the dataset for visualization
df_head_10 = df.head(10)

numerical_columns = ['math_score', 'reading_score', 'writing_score']
for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=df_head_10.index, y=column, data=df_head_10)
    plt.title(f'{column.capitalize().replace("_", " ")} in First 10 Rows')
    plt.xlabel('Index')
    plt.ylabel('Score')
    plt.show()
```
