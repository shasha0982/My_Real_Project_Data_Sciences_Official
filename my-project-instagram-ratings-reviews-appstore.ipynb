Calculate Overall Score for Excellent And Bad Reviews

```python
df = apple

# Define thresholds for good and terrible reviews
good_threshold = 5  # Reviews with a score of 4 or higher are considered good
terrible_threshold = 3  # Reviews with a score of 2 or lower are considered terrible

# Filter reviews based on the defined thresholds
good_reviews = df[df['score'] >= good_threshold]
terrible_reviews = df[df['score'] <= terrible_threshold]

# Calculate overall scores
overall_good_score = good_reviews['score'].mean()
overall_terrible_score = terrible_reviews['score'].mean()

print("Overall Score for Good Reviews:", overall_good_score)
print("Overall Score for Terrible Reviews:", overall_terrible_score)
```

Sentiment Analysis

```python
import pandas as pd
from textblob import TextBlob

# Function to calculate sentiment polarity
def calculate_sentiment(text):
    return TextBlob(text).sentiment.polarity

# Apply the function to the review text
apple['sentiment_polarity'] = apple['text'].apply(calculate_sentiment)

apple.head()
```

```python
# Calculate average sentiment polarity for each score
average_sentiment_by_score = apple.groupby('score')['sentiment_polarity'].mean()

# Plot the results
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(average_sentiment_by_score.index, average_sentiment_by_score.values, color='skyblue')
plt.xlabel('Review Score')
plt.ylabel('Average Sentiment Polarity')
plt.title('Average Sentiment Polarity by Review Score')
plt.xticks(average_sentiment_by_score.index)
plt.grid(axis='y')
plt.show()
```

The distribution of review scores

```python
# Analyze the distribution of review scores
score_distribution = apple['score'].value_counts().sort_index()

# Plot the distribution
plt.figure(figsize=(10, 6))
plt.bar(score_distribution.index, score_distribution.values, color='lightgreen')
plt.xlabel('Review Score')
plt.ylabel('Number of Reviews')
plt.title('Distribution of Review Scores')
plt.xticks(score_distribution.index)
plt.grid(axis='y')
plt.show()
```

The number of reviews per country

```python
# Calculate the number of reviews per country
reviews_per_country = apple.groupby('country').size()

# Plot the number of reviews per country
plt.figure(figsize=(10, 6))
reviews_per_country.plot(kind='bar', color='orange')
plt.xlabel('Country')
plt.ylabel('Number of Reviews')
plt.title('Number of Reviews per Country')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
