```python
# Case Fatality Rate (CFR) by Country
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

covid_filled = pd.read_csv('/kaggle/working/globalstatisticcovid19.csv')

# Convert columns to numeric, removing commas and converting strings to NaN where conversion is not possible
columns_to_convert = ['Total Cases', 'Total Deaths', 'Total Recovered', 'Total Tests', 'Population', 'Active Cases']
for column in columns_to_convert:
    covid_filled[column] = pd.to_numeric(covid_filled[column].str.replace(',', ''), errors='coerce')
    

# Calculate the Case Fatality Rate (CFR) as a percentage
covid_filled['CFR'] = (covid_filled['Total Deaths'] / covid_filled['Total Cases']) * 100

# Plotting
plt.figure(figsize=(14, 8))
sns.barplot(x='Country', y='CFR', data=covid_filled.nlargest(10, 'CFR'), palette='cool')
plt.title('Case Fatality Rate (CFR) by Country')
plt.xlabel('Country')
plt.ylabel('CFR (%)')
plt.xticks(rotation=70)
plt.tight_layout()
plt.show()
```

```python
# Recovery Rate by Country
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the Recovery Rate as a percentage
covid_filled['Recovery Rate'] = (covid_filled['Total Recovered'] / covid_filled['Total Cases']) * 100

# Plotting
plt.figure(figsize=(14, 8))
sns.barplot(x='Country', y='Recovery Rate', data=covid_filled.nlargest(10, 'Recovery Rate'), palette='spring')
plt.title('Recovery Rate by Country')
plt.xlabel('Country')
plt.ylabel('Recovery Rate (%)')
plt.xticks(rotation=80)
plt.tight_layout()
plt.show()
```

```python
# Active Cases per Million by Country
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming covid_filled is your DataFrame and it needs conversion for 'Active Cases' and 'Population'

# Convert 'Active Cases' and 'Population' columns to numeric, handling any non-numeric values
covid_filled['Active Cases'] = pd.to_numeric(covid_filled['Active Cases'].str.replace(',', ''), errors='coerce')
covid_filled['Population'] = pd.to_numeric(covid_filled['Population'].str.replace(',', ''), errors='coerce')

# Calculate Active Cases per Million
covid_filled['Active Cases per Million'] = (covid_filled['Active Cases'] / covid_filled['Population']) * 1e6

# Plotting
plt.figure(figsize=(10, 6))
# Using horizontal bar plot for better readability
sns.barplot(x='Active Cases per Million', y='Country', data=covid_filled.nlargest(10, 'Active Cases per Million'), palette='autumn')

plt.title('Active Cases per Million by Country')
plt.xlabel('Active Cases per Million')
plt.ylabel('Country')

plt.tight_layout()
plt.show()
```

```python
# Tests per Case by Country
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate Tests per Case
covid_filled['Tests per Case'] = covid_filled['Total Tests'] / covid_filled['Total Cases']

# Plotting
plt.figure(figsize=(14, 8))
sns.barplot(x='Country', y='Tests per Case', data=covid_filled.nlargest(10, 'Tests per Case'), palette='winter')
plt.title('Tests per Case by Country')
plt.xlabel('Country')
plt.ylabel('Tests per Case')
plt.xticks(rotation=70)
plt.tight_layout()
plt.show()
```
