```python
import matplotlib.pyplot as plt
import seaborn as sns

data = data_clean_unique

# Set up the visualization layout and aesthetics
sns.set(style="whitegrid")
plt.figure(figsize=(12, 18))

# Create subplots for key categorical variables
plt.subplot(3, 2, 1)
sns.countplot(y='Gender', data=data, order = data['Gender'].value_counts().index)
plt.title('Gender Distribution')

plt.subplot(3, 2, 2)
sns.countplot(y='treatment', data=data)
plt.title('Treatment Distribution')

plt.subplot(3, 2, 3)
sns.countplot(y='family_history', data=data)
plt.title('Family History of Mental Health Conditions')

plt.subplot(3, 2, 4)
sns.countplot(y='Growing_Stress', data=data)
plt.title('Growing Stress Reports')

plt.subplot(3, 2, 5)
sns.countplot(y='Days_Indoors', data=data, order = data['Days_Indoors'].value_counts().index)
plt.title('Days Indoors Distribution')

plt.subplot(3, 2, 6)
sns.countplot(y='Mental_Health_History', data=data)
plt.title('Self-reported Mental Health History')

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Make a copy of the DataFrame for modifications to avoid the SettingWithCopyWarning
data_encoded = data.copy()

# Columns to encode (binary Yes/No columns)
binary_columns = ['self_employed', 'family_history', 'treatment', 'Growing_Stress', 
                  'Changes_Habits', 'Mental_Health_History', 'Coping_Struggles', 
                  'Work_Interest', 'Social_Weakness', 'mental_health_interview']

# Encode binary columns
for col in binary_columns:
    data_encoded[col] = label_encoder.fit_transform(data_encoded[col])

# Encode 'Days_Indoors' with a custom mapping due to its ordinal nature
days_indoors_mapping = {'1-14 days': 1, '15-30 days': 2, 'More than 30 days': 3}
data_encoded['Days_Indoors'] = data_encoded['Days_Indoors'].map(days_indoors_mapping)

# Compute correlation matrix for the encoded data
correlation_matrix_encoded = data_encoded.select_dtypes(include=['number']).corr()

# Plotting the correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_encoded, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

# Plot histograms for a selection of encoded variables
selected_columns_encoded = ['family_history', 'treatment', 'Growing_Stress', 'Mental_Health_History']
data_encoded[selected_columns_encoded].hist(figsize=(10, 8), bins=15)
plt.suptitle('Histograms of Selected Variables')
plt.show()
```

```python
from scipy.stats import chi2_contingency

# Define a function to perform Chi-square test and print results
def perform_chi2_test(df, col1, col2):
    # Creating a contingency table
    contingency_table = pd.crosstab(df[col1], df[col2])
    # Performing the Chi-square test
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    # Printing results
    print(f"Chi-square test between {col1} and {col2}:")
    print(f"Chi-square statistic: {chi2}")
    print(f"P-value: {p}")
    print(f"Degrees of freedom: {dof}")
    print("\n")

# Perform Chi-square tests for specified relationships
perform_chi2_test(data, 'family_history', 'treatment')
perform_chi2_test(data, 'Growing_Stress', 'treatment')
perform_chi2_test(data, 'Mental_Health_History', 'treatment')
```

The results from the Chi-square tests of independence show significant associations between the variables tested and the likelihood of seeking treatment for mental health issues:

* Family History and Treatment: With a chi-square statistic of approximately 39329.82 and a p-value of 0.0, there's a very strong association between having a family history of mental health issues and seeking treatment. This suggests that individuals with a family history of mental health problems are significantly more likely to seek treatment.

* Growing Stress and Treatment: The chi-square statistic is approximately 31.75 with a p-value of about 1.27e-07, indicating a significant relationship between growing stress and seeking treatment. This means that the level of reported stress is associated with the likelihood of seeking treatment for mental health issues.

* Self-reported Mental Health History and Treatment: The chi-square statistic is approximately 13.56 with a p-value of 0.0011, showing a significant association between having a self-reported mental health history and seeking treatment. This suggests that individuals who acknowledge a history of mental health issues are more likely to seek treatment.

* These results highlight the importance of family history, stress levels, and self-recognition of mental health issues in the context of seeking treatment. They underscore the complex interplay of personal, familial, and psychological factors in the pursuit of mental health support.
