```python
# upload new csv files with sql 

df_1 = pd.read_csv('/kaggle/input/add-on-time/2024-03-27T08_21_2400_00_edh6.csv')

# Convert 'date' column to datetime format explicitly
df_1['date'] = pd.to_datetime(df_1['date'])

# Extract year and month as a string for grouping
df_1['year_month_str'] = df_1['date'].dt.strftime('%Y-%m')

# Count the number of reviews per year-month string
df_reviews_time_str = df_1.groupby('year_month_str').size().reset_index(name='counts')

# Plot
plt.figure(figsize=(15, 7))
sns.lineplot(data=df_reviews_time_str, x='year_month_str', y='counts')
plt.title('Number of Reviews Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45)
plt.show()
```

Trends in the number of reviews over time.

```python
# Distribution of scores
plt.figure(figsize=(10, 6))
sns.countplot(x='score', data=df_1)
plt.title('Distribution of Scores')
plt.xlabel('Score')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of reviews by country
plt.figure(figsize=(10, 6))
sns.countplot(y='country', data=df_1, order = df_1['country'].value_counts().index)
plt.title('Distribution of Reviews by Country')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```

```python
# Calculate the length of the review text
df_1['text_length'] = df_1['text'].apply(len)

# Group by score and calculate the average text length for each score
df_score_text_length = df_1.groupby('score')['text_length'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='score', y='text_length', data=df_score_text_length)
plt.title('Average Length of Review Text by Score')
plt.xlabel('Score')
plt.ylabel('Average Text Length')
plt.show()
```

The relationship between the review scores and the length of the review text, providing insights into user engagement and feedback depth across different satisfaction levels.

The relationship between the review scores and the countries in correlation heatmap

```python
# Group by country and score, then count the number of reviews for each combination
df_country_score = df_1.groupby(['country', 'score']).size().reset_index(name='counts')

# Pivot the table to have countries as rows and scores as columns
pivot_df_country_score = df_country_score.pivot(index='country', columns='score', values='counts').fillna(0)

# Plotting
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_df_country_score, annot=True, fmt='.0f', cmap='YlGnBu', linewidths=.5)
plt.title('Number of Reviews by Country and Score')
plt.xlabel('Score')
plt.ylabel('Country')
plt.show()
```
