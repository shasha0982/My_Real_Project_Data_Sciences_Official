```python
# Read the WEOOct2023all.csv file
df_1 = pd.read_csv('/kaggle/input/rural-development-dataset/Global data from the International Monetary Fund (IMF)/WEOOct2023all.csv')

# Print the first 5 rows of the dataframe
df_1.head()
```

```python
# Remove rows with any NaN values
df = df_1.dropna()
```

```python
df
```

```python
# Replacing non-numeric values with NaN
df['Value'] = pd.to_numeric(df['Value'], errors='coerce')

# Display the dataframe again
df
```

```python
# Ensuring that the 'Year' column is of integer type
df['Year'] = df['Year'].astype(int)

# Now repeating the previous operations
gdp_ppp = df[df['Subject Descriptor'] == 'Gross domestic product, current prices']
gdp_ppp = gdp_ppp[(gdp_ppp['Year'] >= 1980) & (gdp_ppp['Year'] <= 2023)]
total_gdp_ppp = gdp_ppp.groupby('Country')['Value'].sum().reset_index()
```

```python
# Let's preview the unique subject descriptors to find the closest match to 'General government revenue'
df['Subject Descriptor'].unique()
```

General government revenue' from 1980 to 2023 and Visualize the top 10 countries

```python
# Selecting General government revenue data
gov_revenue = df[df['Subject Descriptor'] == 'General government revenue']

# Selecting the years between 1980 and 2023
gov_revenue = gov_revenue[(gov_revenue['Year'] >= 1980) & (gov_revenue['Year'] <= 2023)]

# Getting the total government revenue for each country over the years 1980-2023
total_gov_revenue = gov_revenue.groupby('Country')['Value'].sum().reset_index()

# Selecting the top 10 countries based on total government revenue
top10_gov_revenue = total_gov_revenue.nlargest(10, 'Value')

# Plotting the data
plt.figure(figsize=(10, 7))
sns.barplot(x='Value', y='Country', data=top10_gov_revenue, palette='viridis')
plt.xlabel('Total Government Revenue from 1980 to 2023 (in billions)')
plt.ylabel('Country')
plt.title('Top 10 Countries based on their General Government Revenue from 1980 to 2023')
plt.show()
```

The total population across the years 1980-2023 for each country 

```python
# Selecting 'Population' data

population = df[df['Subject Descriptor'] == 'Population']

# Selecting the years between 1980 and 2023

population = population[(population['Year'] >= 1980) & (population['Year'] <= 2023)]

# Getting the total population for each country over the years 1980-2023

total_population = population.groupby('Country')['Value'].sum().reset_index()

# Selecting the top 10 countries based on total population

top10_population = total_population.nlargest(10, 'Value')

# Plotting the data

plt.figure(figsize=(10, 7))
sns.barplot(x='Value', y='Country', data=top10_population, palette='viridis')
plt.xlabel('Total Population from 1980 to 2023 (in billions)')
plt.ylabel('Country')
plt.title('Top 10 Countries based on their Total Population from 1980 to 2023')
plt.show()
```

```python
# Replacing '..' with 'NaN' and converting columns to float
for col in rural.columns[1:]:
    rural[col] = rural[col].str.replace('..', 'NaN').str.replace(',', '').astype(float)

# Replacing NaN values with 0
df = rural.fillna(0)

# Checking the first few rows after conversion
df.head()
```

Data Visualization For Rural Population

```python
# Setting the style
sns.set_style("whitegrid")

# Creating the figure and the axes
fig, ax = plt.subplots(figsize=(15,10))

# Plotting data for the first 5 countries
for country in df[' Country'].head(5):
    ax.plot(df.columns[1:], df[df[' Country'] == country].values[0][1:], label=country)

# Setting xticks, xlabel, ylabel and title
plt.xticks(rotation=90)
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Yearly data for the first 5 countries')

# Adding legend
plt.legend()

# Displaying the plot
plt.show()
```
