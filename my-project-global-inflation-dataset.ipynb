```python
import matplotlib.pyplot as plt
import seaborn as sns

# Choose a country for the time series plot. Replace 'United States' with your country of interest.
selected_country = 'United States'

# Time series plot for the selected country
country_data = data_cleaned[data_cleaned['country_name'] == selected_country]

# Ensure there is data for the selected country to plot
if not country_data.empty:
    plt.figure(figsize=(14, 7))
    years = country_data.columns[2:]  # Assuming the first two columns are 'country_name' and 'indicator_name'
    inflation_rates = country_data.iloc[0, 2:].values
    plt.plot(years, inflation_rates, marker='o')
    plt.title(f'Inflation Rate Trend for {selected_country}')
    plt.xlabel('Year')
    plt.ylabel('Inflation Rate (%)')
    plt.xticks(rotation=45)
    plt.show()
else:
    print(f"No data available for {selected_country}.")
```

```python
# Average Inflation Rate by Country (1980-2023)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Since we want to plot all countries' inflation rates in a bar graph, and considering the extensive range of the dataset,
# a comprehensive view might not be feasible in a single plot. Instead, we can showcase an example with the latest year available
# or an average inflation rate across all years for each country.

# Calculate the average inflation rate across all years for each country
data_cleaned['Average Inflation Rate'] = data_cleaned.iloc[:, 2:-1].mean(axis=1)

# Sort the data by average inflation rate for better visualization
sorted_data = data_cleaned.sort_values(by='Average Inflation Rate')

plt.figure(figsize=(14, 50))
sns.barplot(x='Average Inflation Rate', y='country_name', data=sorted_data, palette='viridis')
plt.title('Average Inflation Rate by Country (1980-2023)')
plt.xlabel('Average Inflation Rate (%)')
plt.ylabel('Country')
plt.show()
```

```python
# Correlation between Annual Inflation Rates Over Years
```

```python
# Calculating the correlation matrix for the cleaned data, excluding 'country_name', 'indicator_name', and 'Average Inflation Rate'
correlation_matrix = data_cleaned.iloc[:, 2:-1].corr()  # Selecting only the years for correlation

# Creating a more readable heatmap visualization
plt.figure(figsize=(20, 15))  # Adjusting the figure size for better readability
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=.5, cbar_kws={'shrink': .5})
plt.xticks(rotation=45, ha='right')  # Rotating x-axis labels for better visibility
plt.yticks(rotation=0)  # Keeping y-axis labels horizontal for readability
plt.title('Correlation between Annual Inflation Rates Over Years', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Year', fontsize=12)
plt.tight_layout()  # Adjusts subplot params so that the subplot(s) fits in to the figure area
plt.show()
```
